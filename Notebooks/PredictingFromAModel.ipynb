The following notebook is a generalised model for carrying out inferences of a trained model. It can identify any uploaded model based on the '.pkl' file extension. Then, as this is an audio model, it will pass any file with the '.wav' extension for predictions. To adjust this for other datatypes the user should enter the relevant file extension.

Set-Up Learner From Uploaded Model

```
path = Path()
extension = str(input("Enter the file extension for prediction: "))
if extension[0]!='.':
  extension = "." + extension
model = (path.ls(file_exts = '.pkl'))
data = (path.ls(file_exts=extension))
print(data[0])
print(model[0])
```

```
learn_inf = load_learner(path/model[0], cpu=True)
```

Runs a loop on the directory and predicts every file uploaded for testing 

```
for i in range(0,len(data)):
  file_path = '/content/' + str(data[i])
  audio = AudioTensor.create(file_path)
  pred,pred_idx,probs = learn_inf.predict(audio)
  value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'
  print(str(data[i]) +" " + value)
```
